```python
import cv2
from matplotlib import pyplot as plt
from skimage.filters import (threshold_otsu, threshold_niblack,
                             threshold_sauvola, try_all_threshold, threshold_local)
import numpy as np
```

```python
path = "E:\\Sem5\\Raw\\Plate_data\\Old\\2.jpg"
frame = cv2.imread(path)
print(frame.shape)
# print(frame)
```

```python
# cv2.imshow('a',frame)
# cv2.waitKey()
```

```python
# path = "C:\\Users\\ASUS\\OneDrive\\Desktop\\download (4).jpg"

# frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

# cv2.resize

plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.imshow(frame, aspect='auto')
plt.title("Origin")

frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
print(frame.shape)


plt.subplot(1,3,2)
plt.imshow(frame, cmap='gray', aspect='auto')
plt.title("Origin in gray")

# frame_blurred = cv2.GaussianBlur(frame, (5, 5), 0)

# plt.subplot(1, 3, 3)
# plt.imshow(frame_blurred, cmap='gray',aspect='auto')
# plt.title("Blurred")
plt.show()
```

```python
width = frame.shape[1]
height = frame.shape[0]
```

```python
_,otsu = cv2.threshold(frame, 0, 255, cv2.THRESH_OTSU)
```

```python
window_size = 35
```

```python
thres_niblack = threshold_niblack(frame,window_size = window_size,k = 0.5)
# print(niblack_image.shape)
niblack_image = frame > thres_niblack
```

```python
thres_sauvola = threshold_sauvola(frame,window_size = window_size)
sauvola_image = frame > thres_sauvola
```

```python
thres_otsu = threshold_otsu(frame)
otsu_image = frame > thres_otsu
```

```python
block_size = 15
local_thresh = threshold_local(frame, block_size, offset=10)
binary_local = frame > local_thresh
```

```python
plt.figure(figsize=(12,8))

plt.subplot(3, 2, 1)
plt.imshow(frame, cmap='gray', aspect='auto')
plt.title('Origin')
plt.axis('off')

plt.subplot(3,2,2)
plt.imshow(otsu, cmap='gray', aspect='auto')
plt.title('otsu cv2')
plt.axis('off')

plt.subplot(3,2,3)
plt.imshow(otsu_image, cmap='gray', aspect='auto')
plt.title('Otsu scikit')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.imshow(niblack_image, cmap='gray', aspect='auto')
plt.title('Niblack')
plt.axis('off')

plt.subplot(3, 2, 5)
plt.imshow(sauvola_image, cmap='gray',aspect='auto')
plt.title('Sauvola')
plt.axis('off')

plt.subplot(3,2,6)
plt.imshow(binary_local, cmap='gray',aspect='auto')
plt.title("Binary local")
plt.axis('off')



plt.show() 
```
